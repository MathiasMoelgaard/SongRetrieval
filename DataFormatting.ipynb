```python
ps = PorterStemmer() 
def tokenizer(text):
    return [word for word in re.findall(r'[a-z]+', text.lower())][0]
```

```python
desiredList = ["happy", "sad", "calm", "energetic"]
removeList = []

dictionary=PyDictionary()

def compare(word1, word2):
    if ps.stem(word1) == ps.stem(word2):
        return True
    for syn in wordnet.synsets(word1):
        for l in syn.lemmas():
            if ps.stem(l.name()) == ps.stem(word2):
                return True
    for word in dictionary.synonym(word1):
        if word == word2:
            return True
    return False

def compareBad(word1, word2):
    if ps.stem(word1) == ps.stem(word2):
        return True
    for syn in wordnet.synsets(word1):
        for l in syn.lemmas():
            if ps.stem(l.name()) == ps.stem(word2):
                return True
    return False

for word in reduced_df.columns:
    for desired in desiredList:
        if compare(tokenizer(desired), tokenizer(word)):
            if desired not in reduced_df.columns:
                reduced_df[desired] = reduced_df[word]
            else:
                reduced_df[desired] = reduced_df[desired] + reduced_df[word]
            removeList.append(word)

print(removeList)
for word in set(removeList):
    if word not in desiredList:
        del reduced_df[word]
reduced_df

cols = list(reduced_df)
for desired in desiredList:
    temp = reduced_df[desired]
    del reduced_df[desired]
    reduced_df.insert(6, desired, temp)

reduced_df
```

```python
removeList = []
for word in reduced_df.columns:
    for syn in reduced_df.columns:
        if word != syn:
            if compare(tokenizer(word), tokenizer(syn)) and word not in removeList:
                sumCol = reduced_df[word] + reduced_df[syn]
                reduced_df[word] = sumCol
                removeList.append(syn)
for word in set(removeList):
    if word not in desiredList:
        del reduced_df[word]
reduced_df
#print(removeList)
#print(set(removeList))
```
